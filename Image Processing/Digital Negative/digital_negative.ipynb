```python
import cv2 as cv2
import matplotlib.pyplot as plt

o_img = cv2.imread('./1.webp',1)
plt.imshow(o_img)
plt.show()

new_img = o_img

height,width,_ = o_img.shape

for i in range(0,height-1):
    for j in range(0,width-1):
        pixel = o_img[i][j]
        
        pixel[0] = 255 - pixel[0]
        pixel[1] = 255 - pixel[1]
        pixel[2] = 255 - pixel[2]

        new_img[i][j] = pixel

cv2.imwrite('./negative_image.png',new_img)


plt.imshow(new_img)
plt.show()
```
